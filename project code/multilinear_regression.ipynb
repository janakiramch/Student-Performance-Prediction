```python
import numpy as np 
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as seabornInstance  
from sklearn import metrics
%matplotlib inline
```

```python
from sklearn.metrics import mean_squared_error, r2_score
```

```python
columns=['year_2','year_3','year_4','tenth','twelve','higher_education','take_notes','prefer_tostudy','hoursstudy','time_relaxing','backlog','backlog_history','participate_activities','attendance','branch_ECE','branch_EEE','branch_EIE','branch_MEE','branch_Other','gender_Male','type_H','studymode_Class notes, Textbook','studymode_Online Tutorials','studymode_Online Tutorials, Class notes','studymode_Online Tutorials, Class notes, Textbook','studymode_Textbook','studymode_Online Tutorials, Textbook']
```

```python
X=data[columns]
Y=data['cgpaup']
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split( X, Y, test_size=0.20, random_state=0)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
regressor = LinearRegression()
```

```python
regressor.fit(X_train,Y_train)
```

```python
Y_pred= regressor.predict(X_test)
Y_pred
```

```python
print('Mean Absolute Error:', metrics.mean_absolute_error(Y_test, Y_pred))  
print('Mean Squared Error:', metrics.mean_squared_error(Y_test, Y_pred))  
print('Root Mean Squared Error:', np.sqrt(metrics.mean_squared_error(Y_test, Y_pred)))
```

```python
from sklearn.model_selection import cross_val_score
```

```python
# 10-fold cross-validation with all  features
lm = LinearRegression()
scores = cross_val_score(lm, X, Y, cv=10, scoring='neg_mean_squared_error')
print(-scores)
```

```python
mse_scores=(-scores)
```

```python
# convert from MSE to RMSE
rmse_scores = np.sqrt(mse_scores)
print(rmse_scores)
```

```python
print("Cross Validated  MSE: %f" % (mse_scores.mean()))
print("Cross Validated RMSE: %f" % (rmse_scores.mean()))
```
