```python
X=data[['year_2','year_3','year_4','tenth','twelve','higher_education','take_notes','prefer_tostudy','hoursstudy','time_relaxing','active_backlogs','backlog_history','participate_activities','attendance','branch_ECE','branch_EEE','branch_EIE','branch_MEE','branch_Other','gender_Male','type_H','studymode_Class notes, Textbook','studymode_Online Tutorials','studymode_Online Tutorials, Class notes','studymode_Online Tutorials, Class notes, Textbook','studymode_Textbook','studymode_Online Tutorials, Textbook']]
Y=data['cgpa']
```

```python
data_dmatrix = xgb.DMatrix(data=X,label=Y)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=0)
```

```python
xg_reg = xgb.XGBRegressor(objective ='reg:linear', colsample_bytree = 0.3, learning_rate = 0.1,
                max_depth = 5, alpha = 10, n_estimators = 10)
```

```python
xg_reg.fit(X_train,Y_train)
preds = xg_reg.predict(X_test)
```

```python
rmse = np.sqrt(mean_squared_error(Y_test, preds))
print("RMSE: %f" % (rmse))
```
