```python
from sklearn.ensemble import GradientBoostingRegressor
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error
from sklearn.metrics import mean_absolute_error
```

```python
import matplotlib.pyplot as plt
import seaborn as seabornInstance  
from sklearn import metrics
%matplotlib inline
```

```python
X=data[['year_2','year_3','year_4','tenth','twelve','higher_education','take_notes','prefer_tostudy','hoursstudy','time_relaxing','active_backlogs','backlog_history','participate_activities','attendance','branch_ECE','branch_EEE','branch_EIE','branch_MEE','branch_Other','gender_Male','type_H','studymode_Class notes, Textbook','studymode_Online Tutorials','studymode_Online Tutorials, Class notes','studymode_Online Tutorials, Class notes, Textbook','studymode_Textbook','studymode_Online Tutorials, Textbook']]
Y=data['cgpa']
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2,random_state=0)
```

```python
regressor = GradientBoostingRegressor(
    max_depth=4,
    n_estimators=3,
    learning_rate=0.3
)
regressor.fit(X_train, Y_train)
```

```python
Y_pred = regressor.predict(X_test)
```

```python
print('Mean Absolute Error:', metrics.mean_absolute_error(Y_test, Y_pred))  
print('Mean Squared Error:', metrics.mean_squared_error(Y_test, Y_pred))  
print('Root Mean Squared Error:', np.sqrt(metrics.mean_squared_error(Y_test, Y_pred)))
```

```python
from sklearn.model_selection import cross_val_score
```

```python
regressor = GradientBoostingRegressor(max_depth=4,n_estimators=3,learning_rate=0.3)
scores = cross_val_score(regressor, X, Y, cv=10, scoring='neg_mean_squared_error')
print(-scores)
```

```python
rmse_scores = np.sqrt(-scores)
#print(rmse_scores)
```

```python
print("Cross Validated  MSE: %f" % (-scores.mean()))
print("Cross Validated RMSE: %f" % (rmse_scores.mean()))
```
