```python
import pandas as pd
import numpy as np
```

```python
X=data[['year_2','year_3','year_4','tenth','twelve','higher_education','take_notes','prefer_tostudy','hoursstudy','time_relaxing','active_backlogs','backlog','participate_activities','attendance','branch_ECE','branch_EEE','branch_EIE','branch_MEE','branch_Other','gender_Male','type_H','studymode_Class notes, Textbook','studymode_Online Tutorials','studymode_Online Tutorials, Class notes','studymode_Online Tutorials, Class notes, Textbook','studymode_Textbook','studymode_Online Tutorials, Textbook','cgpaup']]
Y=data['grade']
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.3, random_state=2)
```

```python
from sklearn.ensemble import RandomForestClassifier
classifier = RandomForestClassifier(n_estimators=10,random_state=2)
classifier.fit(X_train, Y_train)
```

```python
Y_pred = classifier.predict(X_test)
```

```python
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score
print(classification_report(Y_test,Y_pred))
print(accuracy_score(Y_test, Y_pred))
```

```python
confusion_matrix(Y_test,Y_pred)
```

```python
import matplotlib.pyplot as plt
import seaborn as sn
plt.figure(figsize=(10,7))
sn.heatmap(confusion_matrix(Y_test,Y_pred), annot=True)
plt.xlabel('Average,Below_average,Excellent,Good,low')
plt.ylabel('Truth')
```

```python
print(accuracy_score(Y_test, Y_pred))
```
